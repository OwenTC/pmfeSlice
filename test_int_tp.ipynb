```python
tp_sig_file = "sig_file.txt"

with open(tp_sig_file) as f:
    for x in f.readlines():
        x = x.strip("\n")
        _,tp_str = x.split(":")
        tps = [eval(t) for t in tp_str.split(";")]
        if len(tps) < 3:
            continue
            
        tp_array = np.array(tps, dtype=int)
        
        #a*x + 0*y + c*z + d*w = E
        #a*x + c*z - E = -d*w
        # A = [x, z, -1], b = [-w1, -w2, -w3, ...]^T
        # Ax = b implies x = [a, c, E]^T
        A = tp_array[:,[0,2]]
        A = np.append(A, -np.ones((1,3), dtype=int).T, axis=1) #[a, c, E]
        b = -tp_array[:,3] #-w
        x = np.linalg.solve(A, b)
        e = 0.01
        if any(~((np.mod(x, 1) < e) | (np.abs(1-np.mod(x, 1)) < e))):
            print(tp_str, x)
                   
```
